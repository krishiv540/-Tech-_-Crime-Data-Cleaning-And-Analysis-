```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load datasets (replace with your actual file paths)
auto_theft_data = pd.read_csv('30_Auto_theft.csv')
rape_victims_data = pd.read_csv('20_Victims_of_rape.csv')

# --- 1. Auto Theft Data: Bar Chart for Stolen vs Recovered Auto Thefts ---

# Grouping by 'Year' and summing up thefts
auto_theft_summary = auto_theft_data.groupby('Year').sum(numeric_only=True)

years = np.array(auto_theft_summary.index)
stolen = auto_theft_summary['Auto_Theft_Stolen'].values
recovered = auto_theft_summary['Auto_Theft_Recovered'].values

plt.figure(figsize=(10, 6))
bar_width = 0.4

# Bar plots for stolen and recovered
plt.bar(years - bar_width/2, stolen, width=bar_width, label='Stolen')
plt.bar(years + bar_width/2, recovered, width=bar_width, label='Recovered')

plt.xlabel('Year')
plt.ylabel('Number of Thefts')
plt.title('Auto Theft: Stolen vs. Recovered by Year')
plt.xticks(years, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# --- 2. Rape Victims Data: Line Chart for Total Cases Over Time ---

# Filter for total rape victims
rape_victims_summary = rape_victims_data[rape_victims_data['Subgroup'] == 'Total Rape Victims']
rape_cases_by_year = rape_victims_summary.groupby('Year')['Rape_Cases_Reported'].sum()

plt.figure(figsize=(10, 6))

# Line plot for trend over time
plt.plot(rape_cases_by_year.index, rape_cases_by_year.values, marker='o', color='red')

plt.xlabel('Year')
plt.ylabel('Total Rape Cases Reported')
plt.title('Trend of Total Rape Cases Reported Over Time')
plt.grid(True)
plt.tight_layout()
plt.show()
```
